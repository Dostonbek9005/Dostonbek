```python
import numpy as np
import pandas as pd
import matplotlib.pyplot  as plt
```

```python
x = np.arange(0, 4*np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)
plt.figure(figsize=(8,4))
# sin(x) grafigi

plt.plot(x, sin, linewidth=1.5,
                linestyle='--',
                color='black',
                marker='h')
plt.title('sin(x) funksiya grafigi')
plt.grid(color='black', linestyle='--')
```
